```python
Fraud = df[df['Class'] == 1]
valid = df[df['Class'] == 0]

frac = len(Fraud)/float(len(valid))
```

```python
print(frac)
print('Fraud Card: {}'.format(len(Fraud)))
print('Valid Card: {}'.format(len(valid)))
```

```python
corr_matrix = df.corr()
fig = plt.figure(figsize = (12,12))

sns.heatmap(corr_matrix,vmax=.8,square=True)
plt.show()
```

```python
col = df.columns.tolist()
col = [c for c in col if c not in ['Class']]

target = 'Class'

x = df[col]
y= df[target]
```

```python
print(x.shape)
```

```python
print(y.shape)
```

```python
from sklearn.metrics import classification_report, accuracy_score
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
```

```python
classifier = {
    'Isolation Forest':IsolationForest(max_samples=len(x),contamination = frac,random_state=1),
    'Local Outlier factor':LocalOutlierFactor(n_neighbors=20,contamination = frac)
}
```

```python
n_outlier = len(Fraud)

for i,(cif_name,cif) in enumerate(classifier.items()):
    
    if cif_name == 'Local Outlier factor':
        y_pred = cif.fit_predict(x)
        score_pred = cif.negative_outlier_factor_
    else:
        cif.fit(x)
        score_pred = cif.decision_function(x)
        y_pred = cif.predict(x)
        
    y_pred[y_pred ==1] = 0
    y_pred[y_pred == -1] = 1
    
    error= (y_pred != y).sum()
```

```python
print('{}:{}'.format(cif_name,error))
print(accuracy_score(y,y_pred))
print(classification_report(y,y_pred))
```
